```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Membaca dataset
df = pd.read_csv('/content/Kelas D_Happiness and Satisfaction.csv', sep=';')

# 2. Membersihkan data
df['GDP'] = df['GDP'].str.replace(',', '.').astype(float)
df['Region'] = df['Region'].str.strip("'\" ")

# 3. Filter Latin America and Caribbean
lac_data = df[df['Region'] == 'Latin America and Caribbean']

# 4. Ambil 5 negara GDP tertinggi
top5_lac = lac_data.nlargest(5, 'GDP')

# 5. Bar Chart
fig, ax = plt.subplots(figsize=(8, 5))

sns.barplot(x='Country', y='GDP', data=top5_lac,
            hue='Country', palette='viridis', legend=False, ax=ax)

ax.set_title('5 Negara dengan Nilai GDP Tertinggi\ndi Latin America and Caribbean',
             fontsize=12, pad=15)
ax.set_xlabel('Negara', fontsize=11)
ax.set_ylabel('GDP', fontsize=11)

# Label nilai di atas setiap bar
for index, row in enumerate(top5_lac.itertuples()):
    ax.text(index, row.GDP + 0.002, f'{row.GDP:.5f}',
            ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.show()
```
