```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Membaca dataset dengan pemisah titik koma (;)
df = pd.read_csv('/content/Kelas D_Happiness and Satisfaction 2.csv', sep=';')

# 2. Membersihkan data string menjadi float (mengganti koma menjadi titik)
df['GDP'] = df['GDP'].str.replace(',', '.').astype(float)
df['Region'] = df['Region'].str.strip("'\" ")

# 3. Memfilter wilayah 'Latin America and Caribbean'
lac_data = df[df['Region'] == 'Latin America and Caribbean']

# 4. Mengambil 5 negara dengan nilai GDP tertinggi
top5_lac = lac_data.nlargest(5, 'GDP')

# 5. Membuat visualisasi Bar Chart
plt.figure(figsize=(8, 5))
sns.barplot(x='Country', y='GDP', data=top5_lac, palette='viridis')
plt.title('5 Negara dengan Nilai GDP Tertinggi di Latin America and Caribbean', fontsize=12, pad=15)
plt.xlabel('Negara')
plt.ylabel('GDP')

# Menambahkan label angka di atas setiap batang grafik
for index, row in enumerate(top5_lac.itertuples()):
    plt.text(index, row.GDP + 0.02, f'{row.GDP:.5f}', ha="center", fontsize=10)

plt.tight_layout()
plt.show()
```
