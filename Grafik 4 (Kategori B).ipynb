```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# ── 1. Load data ────────────────────────────────────────────────
df = pd.read_csv(
    "/content/Kelas D_Happiness and Satisfaction.csv",
    sep=";",
    decimal=","
)
df.columns = df.columns.str.strip()
df["Region"] = df["Region"].str.strip().str.strip("'")
df["adjusted_satisfaction (0-100)"] = pd.to_numeric(
    df["adjusted_satisfaction (0-100)"], errors="coerce"
)
df.dropna(subset=["adjusted_satisfaction (0-100)", "Region"], inplace=True)

# ── 2. Susun data per region, urutkan berdasarkan median ────────
regions = df.groupby("Region")["adjusted_satisfaction (0-100)"]
region_order = (
    regions.median().sort_values(ascending=False).index.tolist()
)
data_per_region = [regions.get_group(r).values for r in region_order]

# ── 3. Warna menggunakan colormap (tanpa seaborn) ────────────────
cmap = plt.cm.get_cmap("Set2", len(region_order))
colors = [cmap(i) for i in range(len(region_order))]

# ── 4. Plot dengan matplotlib murni ─────────────────────────────
fig, ax = plt.subplots(figsize=(14, 7))

bp = ax.boxplot(
    data_per_region,
    patch_artist=True,
    notch=False,
    widths=0.55,
    medianprops=dict(color="white", linewidth=2),
    whiskerprops=dict(linewidth=1.2),
    capprops=dict(linewidth=1.2),
    flierprops=dict(marker="o", markersize=5, alpha=0.5,
                    markeredgewidth=0)
)

# Terapkan warna per box
for patch, color in zip(bp["boxes"], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.85)
for flier, color in zip(bp["fliers"], colors):
    flier.set_markerfacecolor(color)

# Tambahkan jitter plot (strip plot manual)
for i, (vals, color) in enumerate(zip(data_per_region, colors), start=1):
    jitter = np.random.uniform(-0.18, 0.18, size=len(vals))
    ax.scatter(i + jitter, vals, color="black", s=18,
               alpha=0.3, zorder=3, linewidths=0)

# Anotasi nilai median
for i, vals in enumerate(data_per_region, start=1):
    med = np.median(vals)
    ax.text(i, med + 2, f"{med:.0f}",
            ha="center", va="bottom", fontsize=8.5,
            fontweight="bold", color="#333")

# ── 5. Label & estetika ──────────────────────────────────────────
ax.set_xticks(range(1, len(region_order) + 1))
ax.set_xticklabels(region_order, rotation=35, ha="right", fontsize=8.5)
ax.set_title(
    "Perbandingan Sebaran Tingkat Kepuasan (adjusted_satisfaction)\nantar Wilayah Regional",
    fontsize=14, fontweight="bold", pad=16
)
ax.set_xlabel("Region", fontsize=11, labelpad=10)
ax.set_ylabel("Adjusted Satisfaction (0–100)", fontsize=11)
ax.yaxis.grid(True, linestyle="--", alpha=0.5)
ax.set_axisbelow(True)
ax.spines[["top", "right"]].set_visible(False)

plt.tight_layout()
plt.savefig("boxplot_satisfaction_by_region.png", dpi=150, bbox_inches="tight")
plt.show()
```
